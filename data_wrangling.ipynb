```python
plt.figure(figsize=(15, 7))

plt.subplot(1, 2, 1)

sns.histplot(data=df['closnesscentrality'], bins=10, kde=True)
mean = df['closnesscentrality'].mean()
median = df['closnesscentrality'].median()
plt.axvline(mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='purple', linestyle='-.', label=f'Median: {median:.2f}')
plt.xlabel('Closeness Centrality')
plt.xticks(np.arange(0, 1.1, 0.1))
plt.title('Closeness Centrality Histogram')
plt.legend()

plt.subplot(1, 2, 2)

sns.histplot(data=df['harmonicclosnesscentrality'], bins=10, kde=True)
mean = df['harmonicclosnesscentrality'].mean()
median = df['harmonicclosnesscentrality'].median()
plt.axvline(mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
plt.axvline(median, color='purple', linestyle='-.', label=f'Median: {median:.2f}')
plt.xticks(np.arange(0, 1.1, 0.1))
plt.xlabel('Harmonic Closeness Centrality')
plt.title('Harmonic Closeness Centrality Histogram')
plt.legend()

plt.tight_layout()
plt.savefig('Assets/network/closeness_centrality_distr.pdf', dpi=10000, bbox_inches='tight')
plt.show()
```
