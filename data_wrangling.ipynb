```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def sigmoid_function(x):
    return 1 / (1 + np.exp(-(1.85*x - 8.15)))

# Generate x values
x_values = np.linspace(0, 10, 400)

# Compute y values
y_values = sigmoid_function(x_values)

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='1 / (1 + e^(-(1.85*x - 8.15)))')
plt.title('Sigmoid Function Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```
