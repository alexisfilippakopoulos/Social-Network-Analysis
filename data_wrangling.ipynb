### Number of triangles

```python
edges = pd.read_csv('Assets\edge_data.csv')
edges.head(10)
```

```python
G = nx.from_pandas_edgelist(edges, 'Source', 'Target')

# Compute the number of triangles
num_triangles = sum(nx.triangles(G).values()) / 3

print("Number of triangles:", num_triangles)
```

```python
bridges = list(nx.bridges(G))
len(bridges)
```

```python
bridge_nodes = []
    
for bridge in bridges:
    bridge_nodes.append(bridge[0])

bridge_node_counts = Counter(bridge_nodes)

sorted_nodes = sorted(bridge_node_counts.items(), key=lambda x: x[1], reverse=True)

df['bridges'] = 0

print("Number of bridges each node is involved in:")
for node, count in sorted_nodes:
    print(f"{df[df['Id'] == node]['Label'].item()} is involved in {count} bridges.")
    df.iloc[df[df['Id'] == node].index, -1] = count
    
```

### Topic Details Homophily

```python
def clean_and_split_topic(topic):
    # Remove content inside parentheses
    topic = re.sub(r'\([^()]*\)', '', topic)
    # Remove the "|" symbol and leading/trailing whitespaces
    topic = topic.replace('|', '').strip()
    # Split the cleaned text by spaces to create a list of individual text values
    topic_list = topic.split()
    return topic_list
```

```python
df['topicdetails'].fillna('', inplace=True)
df['cleaned_and_split_topicdetails'] = df['topicdetails'].apply(clean_and_split_topic)
```
